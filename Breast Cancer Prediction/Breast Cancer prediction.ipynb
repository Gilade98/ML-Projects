# Imports

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

# Exploring the data

```python
data_df = pd.read_csv("Breast Cancer Wisconsin (Diagnostic) Data Set.csv")
data_df.shape
```

# Processing the data

```python
data_df['diagnosis'] = data_df['diagnosis'].map({'M':0,'B':1}) #encoding from categorical to numbers
data_df.head()
```

```python
data_df.isna().sum()
```

```python
data_df.drop(columns = ['id', 'Unnamed: 32'], inplace=True) #removing uselss columns
```

```python
data_df.head()
```

```python
X = data_df.drop(columns = 'diagnosis',axis=1)
Y = data_df['diagnosis']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state =2)
print(x_train.shape, x_test.shape, X.shape)
```

### Standarization

```python
scaler = StandardScaler()
x_train_standard = scaler.fit_transform(x_train)
x_test_standard = scaler.transform(x_test)
```

# Building the model

```python
model = tf.keras.models.Sequential([
    layers.Flatten(input_shape = (30,)),
    layers.Dense(20, activation = 'relu'),
    layers.Dense(2 , activation = 'sigmoid')
])
```

```python
model.compile(optimizer = 'adam',
             loss = 'sparse_categorical_crossentropy',
             metrics = ['accuracy'])
```

```python
epochs = 10
```

```python
history = model.fit(x_train_standard, y_train, validation_split = 0.1, epochs=epochs)
```

# Visualizing the model

```python
plt.plot(history.history['loss'], label = "Train loss")
plt.plot(history.history['val_loss'], label = "Val loss")
plt.title('Loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend()
plt.show()
```

```python
plt.plot(history.history['accuracy'], label = "Train accuracy")
plt.plot(history.history['val_accuracy'], label = "Val accuracy")
plt.title('Accuracy')
plt.xlabel('epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
loss, accuracy = model.evaluate(x_test_standard,y_test)
print("Accuracy is: ", accuracy, "\nLoss: ", loss)
```
